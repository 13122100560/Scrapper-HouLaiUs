```python
sess = requests.Session()
```

```python
def pick_proxy():
    proxy_dict = {}
    index = np.random.choice(proxies.shape[0])
    proxy = proxies['concated'][index]
    head = proxies.loc[index]['type']
    proxy_dict[head] = proxy
    return proxy_dict
```

```python
def scrapper(url):
    req = sess.get(url, cookies=cookies, proxies = pick_proxy())
    if req.status_code == 200:
        page = req.text.replace('\n', '').replace(" ", '')
        commentPatter = re.compile(r'<divclass="comment">.*?<pclass="">(.*?)</p>')
        comment = re.findall(commentPatter, page)
        votesPattern = re.compile(r'<spanclass="votes">(.*?)</span>')
        votes = re.findall(votesPattern, page)
        ratingsPattern = re.compile(r'<spanclass="allstar.*?rating"title="(.*?)"></span>')
        ratings = re.findall(ratingsPattern, page)
        nextPattern = re.compile(r'<ahref="\?(start=\d{2,9}?&amp;limit=20&amp;sort=new_score&amp;status=P&amp);.*?后页>')
        nextUrl = re.findall(nextPattern, page)
        df = pd.DataFrame(columns=['comment', 'votes', 'ratings'])
        df['comment'] = comment
        df['votes'] = votes
        df['ratings'] = ratings
        df.to_csv('houlaiUs.csv', encoding = 'utf-8-sig')
        
        nextPage = 'https://movie.douban.com/subject/26683723/comments?{}'.format(nextUrl[0])
        print("Scrapping : " + nextPage)
        scrapper(nextPage)
```
